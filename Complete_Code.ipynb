```python
df['time']=df['date_time'].apply(lambda x: x.split(' ')[1])
df['date']=df['date_time'].apply(lambda x: x.split(' ')[0])
```

```python
df.drop('date_time',axis=1,inplace=True)
```

```python
df.head()
```

```python
def adjust_time(text):
    text=text.replace('00','')
    text=text.replace(':','')
    text=text.strip('')
    return text
df['time']=df['time'].apply(adjust_time)
```

```python
df=df[df['time']!='']
```

```python
df.shape
```

```python
df['time']=df['time']
df['time']=df['time'].astype('object')
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df['Month']=df['date'].dt.month_name()
```

```python
df['Year']=df['date'].dt.year
df['Year']=df['Year'].astype('object')
```

```python
df['day']=df['date'].dt.day
df['day']=df['day'].astype('object')
```

```python
df.drop('date',axis=1,inplace=True)
```

```python
# Import required libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor,AdaBoostRegressor
from sklearn.metrics import make_scorer, mean_squared_error
```

```python
models={
    'RandomForestRegresser':RandomForestRegressor(),
    'GradientBoostingRegresser':GradientBoostingRegressor(),
    'AdaBoostingRegresser':AdaBoostRegressor(),
    'KNN':KNeighborsRegressor()
}
```

```python
parmas={
    'RandomForestRegresser':{
        'n_estimators':[50,100,150,200],
        'max_depth':[1,3,5,7]
    },
    'GradientBoostingRegresser':{
        'n_estimators':[50,100,150,200],
        'learning_rate':[0.1,0.001,0.001,0.0001]
    },
    'AdaBoostingRegresser':{
        'n_estimators':[50,100,150,200],
        'learning_rate':[0.1,0.001,0.001,0.0001]
    },
    'KNN':{
        'n_neighbors':[1,2,3,4,5,6,7]
    }
}
```

```python
for model_name,model in models.items():
    gr=GridSearchCV(model,parmas[model_name],cv=5,scoring='neg_mean_squared_error')
    gr.fit(X_train,y_train)
    print(f'{model_name}: best score {gr.best_score_} with best params {gr.best_params_}')
```

```python
best_model=GradientBoostingRegressor(n_estimators=200,max_depth=7)
best_model.fit(X_train,y_train)
```

```python
ypre=best_model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,ypre)
```

```python
# Hyperparameter tuning for XGB
# Initialize the XGBoost regressor
xgb_model = XGBRFRegressor(eval_metric='rmse')

# Define the parameter grid to search
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 4, 5],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0]
}
```

```python
# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=xgb_model,
                           param_grid=param_grid,
                           scoring='neg_mean_squared_error',  # For regression
                           cv=5,  # Number of cross-validation folds
                           verbose=1,
                           n_jobs=-1)  # Use all available cores
```

```python
# Fit the model to the training data
grid_search.fit(X_train, y_train)
```

```python
# Get the best parameters and best score
print("Best Parameters: ", grid_search.best_params_)
print("Best Score (Negative MSE): ", grid_search.best_score_)

# Predict on the test set using the best model
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# Calculate and print the test set MSE
test_mse = mean_squared_error(y_test, y_pred)
print("Test Mean Squared Error: ", test_mse)
```

```python
r2_score(y_test,y_pred)
```
